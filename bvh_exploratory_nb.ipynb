Predict forest cover type from the features of the area (Elevation, Slope, Soil, etc.) 

setup enviornment 

```
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px 

```

import and preview the data

```
dataset = pd.read_csv("/content/drive/My Drive/forest-cover-type-prediction/train.csv")
print(dataset.shape)
dataset.head()
```

rename the cover types to useful names

```
dataset['Cover_Type'].replace({1:'Spruce/Fir', 2: 'Lodgepole Pine', 3: 'Ponderosa Pine', 4: 'Cottonwood/Willow', 5: 'Aspen', 6: 'Douglas Fir', 7: ' Krummholz'}, inplace=True)
```

count and print a summary of the cover categories 

```
(cover , cover_counts) = np.unique(dataset.Cover_Type , return_counts=True)
```

```
print('\n The represented cover categories in the data:',cover)
print('\n Each cover type is represented', cover_counts[1], 'times')
```

```
fig =  px.histogram( dataset, x="Cover_Type" , color="Cover_Type")
fig.show()
```

The other categorical columns we have are Soil Types and Wilderness Area

```
dataset.columns
```

rename the wilderness areas to useful names, and then condense Wilderness areas and soil type into one column.

```
dataset = dataset.rename(columns={"Wilderness_Area1":"Rawah_WA","Wilderness_Area2":"Neota_WA",
"Wilderness_Area3":"Comanche_Peak_WA","Wilderness_Area4":"Cache_la_Poudre_WA"})
```

```
dataset['Wilderness_Areas'] = (dataset.iloc[:,11:15] == 1).idxmax(1)
dataset['Soil_Type'] = (dataset.iloc[:,15:55] == 1).idxmax(1)
```

Show the coverage type by these features

```
fig = px.histogram(dataset, x="Wilderness_Areas", color="Cover_Type", barmode="group")
fig.show()
```

```
fig = px.histogram(dataset, x="Soil_Type", color="Cover_Type")
fig.show()
```
